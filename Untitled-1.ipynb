```python
# 2- What is the Havasupai Tribe case about?
The Havasupai Tribe case is when Arizona University’s scientists had used blood samples from members of the tribe for research on type 2 diabetes in unrelated genetic studies. The Tribe did not agree and felt this was a violation of informed consent and then went to handle it through a rough legal battle. 
```

```python
# 3- What was the largest health data breach in history?
The largest health data breach in history may be the 2015 Anthem Inc., a major health insurance provider. This breach affected 78.8 million people as names, addresses, ids, and general private health information, and more were taken and it ended with a $115 million in settlements.
```

```python
# 4# - What is the pulse oximeter controversy about?
The controversy is how for decades, studies have shown that the devices overestimate oxygen levels in patients with darker skin tones and therefore darker skin tones may have false readings.
```

```python
# 5- What is the role of an Institutional Review Board (IRB) in research?
Review research proposals to assess ethical considerations.Ensure informed consent is obtained from participant., Protect the rights and welfare of research subjects ,Monitor ongoing research for compliance with ethical standards. Evaluate risks versus benefits of the research study, and provide guidance on ethical research practices and regulations.
```

```python
# 6- What is the difference between spreadsheet software and database software for health data management?
The difference is how spreadsheet software has columns and rows to organize and show the data, rather than databases being able to use tables with large data.
```

```python
# 7- What is a conflict of interest in public health research?
	A conflict of interest in public health research exists when an investigator’s self-interests have the potential to compromise an investigator's professional judgment and objectivity in the design, conduct or reporting of research.
```

```python
# 8- What is informed consent in the context of health data collection?
Along with transparency, informed consent is the process of obtaining permission from individuals before collecting their health data. 
```

```python
# 9- What is data masking in the context of de-identification of health data?
Data masking is being able to fictionalize the data and or replace it to protect the sensitivity of information.
```

```python
# 10- What is the principle of  K  -anonymity in the context of de-identification of health data?
K-anonymity means that each individual’s data cannot be distinguished from at least k–1 other individuals in a dataset.
```

```python
# 11- What is the main concern about conflicts of interest in public health research?
-The main concern is that conflicts of interest can bias research design, data interpretation, or conclusions which therefore could complicate research.
```

```python
# 12- What is the first step in managing conflicts of interest in public health research?
I believe the first step is full disclosure of any financial, professional, or personal interests that could influence the research.
```

```python
# 13- What is the role of independent oversight in managing conflicts of interest in public health research?
An independent oversight such as ethics committees or institutional review boards could be used to be able to review, monitor, and manage conflicts to ensure research decisions.
```

```python
# 14- What is one strategy for mitigating conflicts of interest in public health research?
One strategy is separating funding sources from research decision making so there is no unbiased or extra motivation for an answer.
```

```python
# 15- What is the potential impact of not properly managing conflicts of interest in public health research?
Failure to manage conflicts of interest can lead to biased findings, loss of credibility, loss of public trust, and even harmful public health policies.

```
